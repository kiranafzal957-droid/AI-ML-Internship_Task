```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('iris')
print("Shape of dataset:", df.shape)
print("Columns:", df.columns)
print("\nFirst 5 rows:")
print(df.head())

# Dataset info
print("\nInfo:")
print(df.info())
print("\nSummary Stats:")
print(df.describe())

# Pairplot
sns.pairplot(df, hue='species')
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()

# Histograms
df.hist(figsize=(10,6), edgecolor='black')
plt.suptitle("Histograms", y=1.02)
plt.show()

# Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplot of Features")
plt.show()
```

# Task 1: Iris Dataset Exploration

## Objective
Explore the Iris dataset to understand relationships between features and species using visualization.

##  Dataset Details
 150 rows and 5 columns
 Features: sepal_length, sepal_width, petal_length, petal_width
 Target: species (Setosa, Versicolor, Virginica)

##  Key Visualizations
 **Pairplot**: Shows clear separation of species, especially by petal length/width.
 **Histograms**: Show the distribution of feature values.
 **Boxplots**: Help identify outliers in the dataset.

##  Summary
The dataset is clean (no missing values). Petal measurements are key features for species separation. The dataset is ideal for building simple classification models.
